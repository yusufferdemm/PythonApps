##### Her Videonun Başlık Uzunluğunu içeren yeni bir sütün ekleyin.

```python
def countTitleCount(title):
    return len(title)
```

```python
video["TitleLen"]=video.title.apply(countTitleCount)
video
```

##### Her bir videoda kullanılan etiket sayısını hesaplayarak, oluşan değerleri yeni bir sütun olarak ekleyin.

Not: Her bir etiket "|" işaretiyle ayrılmaktadır.

```python
def split_tag(tag):
    a=tag.split("|")
    return len(a)
```

```python
video["Tag_count"]=video.tags.apply(split_tag)
video
```
